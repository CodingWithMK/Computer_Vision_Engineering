# Week 1 - Lesson 3 - Matplotlib

````python
import matplotlib.pyplot as plt
````

### 1. ```plt.figure()```

- Defines a new figure object
- **Signature**: ```plt.figure(num=None, figsize=None, dpi=None, facecolor=None, edgecolor=None)```

````python
plt.figure(figsize=(6, 4), dpi=100)
````

### 2. ```plt.plot()```

- Lineplot
- Signature: ```plt.plot(x, y, fmt='', label=None, **kwargs)```

````python
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, linestyle='--', marker='o', label='sin')
````

### 3. ```plt.imshow()```

- Displaying 2D arrays/images
- **Signature**: ```plt.imshow(X, cmap=None, interpolation=None, vmin=None, vmax=None)```

````python
color = plt.imread("14-Lamborghini-Revuelto-review-UK-2024.jpg")
plt.imshow(color, cmap='gray', vmin=0, vmax=255)
````

````python
gray = plt.imread("gray.png")
plt.imshow(gray, cmap='gray')
plt.colorbar()
````

````python
gray1 = plt.imread("gray.png")
plt.imshow(gray1, cmap='gray', vmin=0, vmax=255)
plt.colorbar()
````

### 4. ```plt.title()```, ```plt.xlabel()```, ```plt.ylabel()```

- Title and axis labels
- **Signatures**:
    - ```plt.title(label)```
    - ```plt.xlabel(label)```
    - ```plt.ylabel(label)```    

````python
gray = plt.imread("gray.png")
plt.imshow(gray, cmap='gray')
plt.colorbar()

plt.title("Gray Image")
plt.xlabel("X pixels")
plt.ylabel("Y pixels")
````

### 5. ```plt.subplot()```

- Multiple subplot panes.
- **Signature**: ```plt.subplot(nrows, ncols, index, **kwargs)```

````python
img_color = plt.imread("14-Lamborghini-Revuelto-review-UK-2024.jpg")
````

````python
plt.figure()
plt.subplot(1, 2, 1)
plt.imshow(img_color)

plt.subplot(1, 2, 2)
plt.imshow(gray, cmap='gray')
````

### ```plt.show()```

- Displays all available graphs and images

````python

plt.figure()
plt.subplot(1, 2, 1)
plt.imshow(img_color)

plt.subplot(1, 2, 2)
x = np.linspace(0, 10, 100)
y = np.sin(x)

x1 = np.linspace(0, 10, 100)
y1 = np.cos(x)

plt.plot(x, y, linestyle='--', marker='o', label='sin')
plt.plot(x1, y1, linestyle='-', marker='s', label='cos')

plt.legend()
plt.show()
````
